```julia
using Plots
using StatsPlots
using Feather
using DataFrames
d = nothing
```

```julia
d_ed = Dict{Int64, Vector{Float64}}()
for ix in 1:10
    base_path = "/Users/jdlara/Desktop/test-ED-$(ix)/"
    folder = [f for f in readdir(base_path) if occursin("test", f)][1]
    full_path = joinpath(base_path, folder, "raw_output", "step-1-stage-2")
    dirs = readdir(full_path)
    log_file = Vector{Float64}(undef, 96)
    idx = 1
    for (ix_i, d) in enumerate(dirs)
        occursin(".", d) && continue
        full_path = joinpath(base_path, folder, "raw_output", "step-1-stage-2", d)
        file_results = Feather.read(joinpath(full_path, "optimizer_log.feather"))
        log_file[idx] = file_results[!, :solve_time][1]
        idx += 1
    end
    d_ed[ix] = log_file
end
```

```julia
d_cvar = Dict{Int64, Vector{Float64}}()
for ix in 1:10
    base_path = "/Users/jdlara/Desktop/test-CvAr-$(ix)/"
    folder = [f for f in readdir(base_path) if occursin("test", f)][1]
    full_path = joinpath(base_path, folder, "raw_output", "step-1-stage-2")
    dirs = readdir(full_path)
    log_file = Vector{Float64}(undef, 96)
    idx = 1
    for (ix_i, d) in enumerate(dirs)
        occursin(".", d) && continue
        full_path = joinpath(base_path, folder, "raw_output", "step-1-stage-2", d)
        file_results = Feather.read(joinpath(full_path, "optimizer_log.feather"))
        log_file[idx] = file_results[!, :solve_time][1]
        idx += 1
    end
    d_cvar[ix] = log_file
end
```

```julia
p = plot(ylimits = (0.0,0.1), 
         legend = nothing, 
          xticks = 1:10, 
         title = "Basic Economic Dispatch Run Times",
         xlabel = "Day Trial",
         ylabel = "Seconds")
for (k, calc) in d_ed
    boxplot!(calc)
end
display(p)
savefig(p, "ed.pdf")
```
